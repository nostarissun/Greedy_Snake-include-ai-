```python
import matplotlib.pyplot as plt
import numpy as np
import re

def read_data_from_file(file_path):
    try:
        data = []
        # 改进的正则表达式：支持科学计数法
        pattern = r'[-+]?(?:\d+\.?\d*|\.\d+)(?:[eE][-+]?\d+)?'  # 匹配整数、小数、科学计数法
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                match = re.search(pattern, line)
                if match:
                    num_str = match.group(0)
                    try:
                        num = float(num_str)
                        data.append(num)
                    except ValueError:
                        print(f"无法转换 '{num_str}'")
        return np.array(data)
    except FileNotFoundError:
        print(f"文件 {file_path} 未找到")
        return None

# 读取数据
loss_data = read_data_from_file('loss.txt')
reward_data = read_data_from_file('reward.txt')
max_reward_data = read_data_from_file('max_reward.txt')
length_data = read_data_from_file('length.txt')

# 绘图
fig, axes = plt.subplots(4, 1, figsize=(15, 20))
titles = ['Loss', 'Reward', 'Max Reward', 'Length']
datasets = [loss_data, reward_data, max_reward_data, length_data]

for ax, title, data in zip(axes, titles, datasets):
    if data is not None:
        ax.plot(data)
        ax.set_title(title)
        ax.set_xlabel('Episode')
        ax.set_ylabel(title)

plt.tight_layout()
plt.show()
```
