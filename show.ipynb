```python
import matplotlib.pyplot as plt
import numpy as np
import re

# 定义一个函数用于读取文件数据
def read_data_from_file(file_path):
    try:
        data = []
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                # 使用正则表达式提取数值
                match = re.search(r'[-+]?\d*\.\d+|\d+', line)
                if match:
                    num_str = match.group(0)
                    try:
                        num = float(num_str)
                        data.append(num)
                    except ValueError:
                        print(f"无法将 '{num_str}' 转换为浮点数。")
                else:
                    try:
                        num = float(line.strip())
                        data.append(num)
                    except ValueError:
                        print(f"无法将 '{line.strip()}' 转换为浮点数。")
        return np.array(data)
    except FileNotFoundError:
        print(f"文件 {file_path} 未找到。")
        return None

# 读取损失、奖励和最大奖励的数据
loss_data = read_data_from_file('loss.txt')
reward_data = read_data_from_file('reward.txt')
max_reward_data = read_data_from_file('max_reward.txt')
# x = []
# x = np.array(x)
# for i in range(len(reward_data)):
#     if i % 100 == 0:
#         x.append(reward_data)

# 创建一个包含三个子图的图形
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# 绘制损失数据的折线图
if loss_data is not None:
    axes[0].plot(loss_data)
    axes[0].set_title('Loss')
    axes[0].set_xlabel('Episode')
    axes[0].set_ylabel('Loss')

# 绘制奖励数据的折线图
if reward_data is not None:
    axes[1].plot(reward_data)
    axes[1].set_title('Reward')
    axes[1].set_xlabel('Episode')
    axes[1].set_ylabel('Reward')

# 绘制最大奖励数据的折线图
if max_reward_data is not None:
    axes[2].plot(max_reward_data)
    axes[2].set_title('Max Reward')
    axes[2].set_xlabel('Episode')
    axes[2].set_ylabel('Max Reward')

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```
